# Compare how often python frameworks/libraries are mentioned in github repositories

<strong> usage: </strong>

The goal is to get an idea of the usage of different python libraries in different application fields of python.

notice that I only count a library when I can get it somehow via the query term from the github api, so this is not an exact count of how many people actually used a certain python library but nevertheless it gives some assessment on the popularity of different python libraries (note: for simplicity I mix the terms libraries and frameworks --> in this notebook I just refer to them both as libraries)

<strong> important: </strong> 

In oder not to hit github api requests limits this sript takes about 20 minutes to finish. (let it run in the background)



## Python Data Science & Machine Learning libraries mentioned part 1

```python
import requests #library to use to make for example HTTP requests.
import json
import time

section_count = []
query_terms = ["NumPy", "Pandas", "Matplotlib", "Seaborn", "SciPy", "Scikit-learn"]
query_terms_count = []

for query_term in query_terms:
    res = requests.get(
        f"https://api.github.com/search/repositories?q={query_term}&sort=stars&order=desc",
    )
    count = res.json().get("total_count", 0)
    query_terms_count.append(count)
    time.sleep(1)  # To avoid hitting the API rate limit

print(query_terms)
print(query_terms_count)
section_count.append(sum(query_terms_count))


import matplotlib.pyplot as plt
plt.bar(query_terms, query_terms_count)
plt.title("Python Data Science & Machine Learning libraries mentioned")
plt.xticks(rotation='vertical', fontsize=8)
plt.show()
time.sleep(180)  # To avoid hitting the API rate limit
```

```python
print(section_count)
```
